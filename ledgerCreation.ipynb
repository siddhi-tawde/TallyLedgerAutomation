```python
# IMPORTING IMPORTANT LIBRARIES
import pandas as pd
import numpy as np 
import pywinauto as pw
import pyautogui as pg
import pymsgbox 
import time
from pywinauto.application import Application
from pywinauto.keyboard import send_keys
import os
```

```python
#READING THE EXCEL INTO DATAFRAME AND STORING IT INTO MAIN
main = pd.read_excel('Bank_Entry_Excel(1).xlsx',sheet_name = "Ledger_Details")
```

```python
checkList = ['ledger_name',
 'ledger_group',
 'ledger_opening_balance_(if_any)',
 'ledger_status']
```

```python
main
```

```python
columnsToCheck = list(main.columns)
```

```python
#main
if checkList == columnsToCheck and len(columnsToCheck)==4:
    print("Excel is OK we can proceed")
    common_paths = [
    r"C:\Program Files\TallyPrime\tally.exe",
    r"C:\Program Files (x86)\Tally\tally.exe",
    r"C:\Tally\tally.exe"
    ]
    def findTallyPath():
        for path in common_paths:
            if os.path.exists(path):
                return path
        return None
    tallyPath = findTallyPath()
    if not tallyPath:
        tallyPath = pymsgbox.prompt(
        "Tally was not found automatically.\n\n"
        "Please enter the full path to the Tally folder or tally.exe.\n\n"
        "✅ Example:\n"
        "C:\\Program Files\\TallyPrime  (we’ll auto-detect tally.exe)")
        if tallyPath and os.path.isdir(tallyPath):
            tallyPath = os.path.join(tallyPath, "tally.exe")

        if not tallyPath or not os.path.exists(tallyPath):
            pymsgbox.confirm("Invalid or no path provided. Exiting.")
            exit()
    tallyPath = os.path.normpath(tallyPath)
    messageCheck = pymsgbox.confirm(f"Tally found at:\n{tallyPath}\nPress ok")
    print(messageCheck)
    if messageCheck=='OK':
        mainApp = Application(backend="win32").start(tallyPath)
        time.sleep(10)
        send_keys('t')  ###remove this in real execution
        time.sleep(4)
        send_keys('{F3}')
        time.sleep(4)
        companySelectionCheck = pymsgbox.confirm('Please select the company!')
        time.sleep(5)
        companyValidationCheck = pymsgbox.confirm('Would you like to continue with selected company?',buttons=['Yes','No'])
        if companyValidationCheck=='No':
            send_keys('{F3}')
            time.sleep(4)
            companySelectionCheck = pymsgbox.confirm('Please select the company!')
        else:
        # Filter rows where ledger_status is 'new'
            new_ledgers = main[main['ledger_status'].str.lower() == 'new'].drop_duplicates()
            # Create dictionary object row-wise
            ledger_dict = {
                index: {
                    "name": row["ledger_name"],
                    "group": row["ledger_group"],
                    
                }
                for index, row in new_ledgers.iterrows()
            }
            send_keys('c')
            time.sleep(1)
            send_keys('ledger')
            time.sleep(1)
            send_keys('{ENTER}')
            time.sleep(2)
            for item in ledger_dict.values():
                send_keys(item['name'])
                time.sleep(1)
                send_keys('{ENTER}')
                time.sleep(1)
                send_keys('{ENTER}')
                time.sleep(1)

                send_keys(item['group'])
                send_keys('%a')
            pymsgbox.confirm('Ledger are created') 


    
        
       
        
        

    
   
else:
    print('Excel is incorrect')
```
